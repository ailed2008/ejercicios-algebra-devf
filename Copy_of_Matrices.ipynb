```python
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets

iris = datasets.load_iris()
```

```python
data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data1
```

```python
matrix=data1.corr()
```

```python
matrixInversa=np.linalg.inv(matrix)
matrixInversa
```
